# 1. Total Patient Volume
# Question: What is the total number of patient records handled by the hospital to date?

SELECT COUNT(PatientID) AS Total_Patient
FROM hospital_management;

# 2. Patient Demographics: Age Profile
# Question: What is the average age of the patient population treated at this facility?

SELECT ROUND(AVG(Age), 2) AS Average_Age
FROM hospital_management;

# 3. Medical Staff Diversity
# Question: How many unique doctors are currently managing patients within the hospital system?


SELECT COUNT(DISTINCT DoctorName) AS Total_Doctors
FROM hospital_management;

# 4. Gross Hospital Revenue (Billing)
# Question: What is the total gross revenue generated by the hospital based on all patient billings?


SELECT SUM(TotalBillAmount) AS Gross_Revenue
FROM hospital_management;

# 5. Realized Revenue (Collections)
# Question: What is the total amount of revenue that has been successfully collected from patients and insurance?

SELECT SUM(AmountPaid) AS Total_Revenue_Collected
FROM hospital_management;

# 6. Financial Risk: Outstanding Receivables
# Question: What is the total outstanding/pending amount currently owed to the hospital?


SELECT SUM(PendingAmount) AS Total_Outstanding_Balance
FROM hospital_management;


# 7. Disease Prevalence Analysis
# Question: Which diseases are most common among the patients admitted to the hospital?

SELECT
    Disease,
    COUNT(PatientID) AS Patient_Count
FROM hospital_management
GROUP BY Disease
ORDER BY Patient_Count DESC;

# 8. Medical Staffing by Specialization
# Question: What is the distribution of unique doctors across different medical specializations?

SELECT
    Specialization,
    COUNT(DISTINCT DoctorID) AS Doctor_Count
FROM hospital_management
GROUP BY Specialization
ORDER BY Doctor_Count DESC;

# 9. Regional Financial Performance (State-wise)
# Question: Provide a financial breakdown by State, showing the total billed amount, total collected (received), and total pending balance.

SELECT
    State,
    SUM(TotalBillAmount) AS Total_Amount,
    SUM(AmountPaid) AS Amount_Received,
    SUM(PendingAmount) AS Amount_Pending
FROM hospital_management
GROUP BY State
ORDER BY Total_Amount DESC;

# 10. Revenue Distribution by Payment Method
# Question: What is the total billing volume generated through each payment mode (e.g., Card, Cash, Insurance, UPI)?

SELECT
    PaymentMode,
    SUM(TotalBillAmount) AS Total_Bill_Amount
FROM hospital_management
GROUP BY PaymentMode
ORDER BY Total_Bill_Amount DESC;


#11. Operational Efficiency: Length of Stay (LOS).
# Question: What is the average duration of stay for each Admission Reason?

SELECT
    AdmissionReason,
    ROUND(AVG(DATEDIFF(STR_TO_DATE(DischargeDate, '%d-%m-%Y'), STR_TO_DATE(AdmissionDate, '%d-%m-%Y'))), 1) AS Avg_Stay_Days
FROM hospital_management
GROUP BY AdmissionReason
ORDER BY Avg_Stay_Days DESC;
